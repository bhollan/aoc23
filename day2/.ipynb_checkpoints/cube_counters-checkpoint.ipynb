```python
# only 12 red cubes, 13 green cubes, and 14 blue cubes?

red_max = 12
green_max = 13
blue_max = 14

color_max = {
    'red':12,
    'green':13,
    'blue':14
}
```

# Part 1

```python
lines = []

for line in open("./input1.txt", "r").readlines():
    game, draws = line.split(": ")
    id = int(game.split(" ")[1])
    possible = True
    for draw in draws.split(";"):
#         print(draw)
        for cqpair in draw.split(","):
#             print(cqpair)
            parts = cqpair.split(" ")
            if len(parts) == 3:
                parts = parts[1:]
            quant = int(parts[0])
            color = parts[1]
            
            for c, m in color_max.items():
                if color == c and quant > m:
                    possible = False
    if possible:
        lines.append(id)
            
#             print(parts)
    
```

```python
import numpy as np
from collections import Counter
```

```python

color_min_root = {
    'red':1,
    'green':1,
    'blue':1
}
```

```python
print(color_min_root)
```

```python
# find the min needed for each color
# find the 'power' of each game
# append the power

lines = []

for line in open("./input1.txt", "r").readlines():
    line = line.rstrip()
    game, draws = line.split(":")
    id = int(game.split(" ")[1])
    color_min = color_min_root.copy()
    print(draws)
    print(color_min, '========before=======')
    for draw in draws.split(";"):
#         print(draw)
        for cqpair in draw.split(","):
            print(cqpair)
            parts = cqpair.split(" ")
            if len(parts) == 3:
                parts = parts[1:]
            quant = int(parts[0])
            color = parts[1]
            next_color_min = color_min.copy()
            for c, m in color_min.items():
#                 print(c, m)
                if color == c and quant > m:
#                     print("PREUPDATE", next_color_min)
#                     print(next_color_min)
                    next_color_min[color] = quant
#                     print("POST-UPDATE", next_color_min)
            color_min = next_color_min.copy()
#     print(color_min, '=======after=======')
    power = np.prod(list(color_min.values()))
#     print(power)
    lines.append(power)
            
#             print(parts)
    
```

```python
"""Find the location of all symbols and load the digits of all 8 neighbors"""
```

```python
def get_naybs(loc, n, m):
    r, c = loc
    naybs = []
    dels = [-1, 0, 1]
    for dr in dels:
        for dc in dels:
            if dr == 0 and dc == 0:
                continue
            new_r = r + dr
            new_c = c + dc
            if new_r < 0 or new_r >= n:
                continue
            if new_c < 0 or new_c >= m:
                continue
            naybs.append((new_r, new_c))
    return naybs
```

```python
lines = []
symbol_locs = []
seen_cells = set()
tally_numbers = []

for row_i, line in enumerate(open("./inputB.txt", "r").readlines()):
    line = line.rstrip()
    lines.append(list(line))
    for col_i, c in enumerate(line):
        if c not in ".1234567890":
            symbol_locs.append((row_i, col_i))
n = len(lines)
m = len(lines[0])
    
```

```python
seen_cells = set()
```

```python
print(seen_cells)
```

```python
def get_num(addr):
    r, c = addr
    num = ''
    while (lines[r][c].isnumeric()):
        c += 1
    c -= 1
    if (r,c) in seen_cells:
        return 0
    seen_cells.add((r,c))
    while (lines[r][c].isnumeric()):
        num = lines[r][c] + num
        c -= 1
    return int(num)
```

```python
for loc in symbol_locs:
    naybs = get_naybs(loc, n, m)
    print("naybs:=========")
    print(naybs)
    for nayb in naybs:
        row = nayb[0]
        col = nayb[1]
        print(row, col)
        print(" seen?  ", len(seen_cells))
        cell = lines[row][col]
        if cell.isnumeric():
            number = get_num(nayb)
            print("Got de NUMBA: ", number)
            tally_numbers.append(number)
            
```
